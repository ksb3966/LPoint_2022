```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib 
matplotlib.rcParams['font.family'] = 'AppleGothic'
from IPython.display import Image 
```

```python
df = pd.read_csv('LPOINT_BIG_COMP_02_PDDE.csv')
df
```

## 유저 ID와 제품 코드로 Feature ID 생성하기

```python
df['feature_id'] = df['cust'] + '_' +df['pd_c']
df = df[['feature_id', 'cust', 'pd_c', 'rct_no', 'chnl_dv', 'cop_c', 'br_c', 'de_dt', 'de_hr', 'buy_am',
       'buy_ct']]
df
```

```python
# 중복 제거
df = df.drop_duplicates()
df
```

### 고객의 해당 상품 주문 횟수 (UP_orders)

```python
data = df.groupby('feature_id').count()
data
```

```python
uporder = data[['buy_ct']]
uporder = uporder.rename(columns = {'buy_ct' : 'UP_orders'})
```

```python
df = pd.merge(df, uporder, on ='feature_id', how = 'outer')
```

```python
df
```

### 고객의 주문목록 평균 상품개수 (UP_average_pos_in_cart)

```python
cart = df.groupby('cust').mean()
cart = cart.reset_index()
cart = cart[['cust', 'buy_ct']]
cart = cart.rename(columns = {'buy_ct':'UP_average_pos_in_cart'})
cart
```

```python
df = pd.merge(df, cart, on ='cust', how = 'outer')
df
```

### 고객의 해당 상품 주문 횟수 / 전체 주문횟수 (Up_reorder_rate)

```python
df
```

```python
reorder_r = df.groupby('cust').count()
reorder_r = reorder_r.reset_index()
reorder_r
```

```python
denominator = reorder_r[['cust', 'buy_ct']]
denominator = denominator.rename(columns = {'buy_ct' : 'denominator'})
denominator
```

```python
reorder_r = df.groupby(['cust', 'pd_c']).count()
reorder_r = reorder_r.reset_index()
reorder_r = reorder_r.rename(columns = {'buy_ct':'numerator'})
numerator = reorder_r[['cust','pd_c', 'numerator']]
```

```python
numerator
```

```python
fraction_df = pd.merge(numerator, denominator, on = 'cust', how = 'outer')
fraction_df['UP_reorder_rate'] = fraction_df['numerator'] / fraction_df['denominator']
fraction_df['feature_id'] = fraction_df['cust'] + '_' + fraction_df['pd_c']
fraction_df
```

```python
df = pd.merge(df, fraction_df[['feature_id','UP_reorder_rate']], on = 'feature_id', how = 'outer')
df
```
